```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from scipy.optimize import curve_fit
```

```python
data1 = pd.read_csv("moni_RC_MUON_09_05.dat", header = None, delim_whitespace = True,
                   names = ["EM","MUON"])
data2 = pd.read_csv("moni_RC_MUON_09_06.dat", header = None, delim_whitespace = True,
                   names = ["EM","MUON"])
data3 = pd.read_csv("data_press.dat", header = None, delim_whitespace = True, names = ["Press"])
time = pd.date_range("2022-09-05", periods = 48, freq = "1h")
```

```python
tiempo = pd.DataFrame(time, columns = ["Fecha"])
data = pd.concat([data1, data2], ignore_index = True)
data_def = pd.concat([tiempo, data, data3], axis = 1)
data_def = data_def.set_index("Fecha")
data_def.head(5)
```

```python
data_def = data_def.drop(data_def.index[32], axis = 0)
data_def
```

```python
int(np.mean(data_def["MUON"]))
```

```python
MUON = data_def["MUON"]
Pres = data_def["Press"]
Pres_mean = np.mean(Pres)
fig, ax1 = plt.subplots(figsize=(11,5))
ax2 = ax1.twinx()

plt.title("Monitoreo de eventos en componente Munica")
ax1.plot(MUON/(3600*0.950), "rs--", label = "Muons")
ax2.plot(Pres, "bs-", label = "Presin")
ax1.set_xlabel("Fecha")
ax1.set_ylabel("Eventos [$s^{-1} m^{-2}$]")
ax1.grid()
ax1.legend(loc = 2)
ax2.set_ylabel("Presin [hPa]")
ax2.set_ylim(765,758)
ax2.legend()
plt.gcf().autofmt_xdate()
#plt.savefig("Moni_RC_MUON.pdf")
plt.show()
```

```python
Pres_mean = np.mean(Pres)
Count_mean = np.mean(MUON)
D_Press = np.array(Pres)-np.ones(len(Pres))*Pres_mean
D_Count = (np.array(MUON)-np.ones(len(MUON))*Count_mean)/(np.ones(len(MUON))*Count_mean)

fig, ax1 = plt.subplots(figsize=(5,5))
plt.title("Counts as function of press")
ax1.plot(D_Press, D_Count, "bo", label = "Counts as function of press")
ax1.set_xlabel("$\Delta P [hPa]$")
ax1.set_ylabel("$\Delta I / I_0$ [a.u.]")
ax1.grid(linestyle = "--", color = "black")
plt.show()
```

```python
def lineal(x,a,b):
    return a*x + b

valores_ini = [1.6,-0.005]

res, cov = curve_fit(lineal, D_Press, D_Count, p0 = valores_ini)
a_fit, b_fit = res
a_err, b_err = np.sqrt(np.diag(cov))

fig, ax1 = plt.subplots(figsize=(5,5))
plt.title("Counts as function of press")
ax1.plot(D_Press, D_Count, "bo")
ax1.plot(D_Press, lineal(D_Press, a_fit, b_fit ), "r-", label = F"Lineal fit, Slope = {a_fit:.4f} $\pm$ {a_err:.4f}")
ax1.legend()
ax1.set_xlabel("$\Delta P [hPa]$")
ax1.set_ylabel("$\Delta I / I_0$ [a.u.]")
ax1.grid(linestyle = "--", color = "black")
plt.show()
```

```python
Count_correct = np.array(D_Press)*a_fit

fig, ax1 = plt.subplots(figsize=(5,5))
plt.title("Counts as function of press")
ax1.plot(D_Press, D_Count, "bo")
ax1.plot(D_Press, lineal(D_Press, a_fit, b_fit ), "r-", label = F"Lineal fit, Slope = {a_fit:.4f} $\pm$ {a_err:.4f}")
ax1.plot(D_Press, Count_correct, "go")
ax1.legend()
ax1.set_xlabel("$\Delta P [hPa]$")
ax1.set_ylabel("$\Delta I / I_0$ [a.u.]")
ax1.grid(linestyle = "--", color = "black")
plt.show()
```
